```python
import pandas
from abnum import Abnum, greek
g = Abnum(greek).value
p = Abnum(greek).preprocess
```

```python
#http://clas-lgpn4.classics.ox.ac.uk:8080/exist/apps/lgpn/browse.html
df = pandas.read_csv("LGPN.csv", sep="\t")
del df['Id']
del df['Update ']
```

```python
df
```

```python
df['Uncial'] = df['Name '].apply(lambda x: p(str(x).upper()))
df.drop_duplicates(subset=['Uncial'], keep='first', inplace=True)
df['Isopsephy'] = df['Uncial'].apply(lambda x: g(x))
```
